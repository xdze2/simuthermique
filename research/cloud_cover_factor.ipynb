```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
cloud_cover = np.linspace(0, 1, 23) 
```

```python
direct = {}

direct['KastenCzeplack'] = 1 - .75*(cloud_cover)**3.4
direct['Rome'] = 1.1 - .757*(cloud_cover)**6
direct['Bologna'] = 1.1 - .99*(cloud_cover)**1.3
#direct['SecondOrderPoly'] = 1.049 + 0.098*cloud_cover - 0.021*cloud_cover**2

for k, v in direct.items():
    plt.plot(cloud_cover, v, label=k)
    
plt.legend(); plt.xlabel('cloud cover');
plt.ylabel('I/I0');
```

```python
diffuse = {}

diffuse['KastenCzeplack'] = .3 + .65*(cloud_cover)**2
diffuse['Rome'] = .7 + .65*(cloud_cover)**2
#direct['SecondOrderPoly'] = 1.049 + 0.098*cloud_cover - 0.021*cloud_cover**2

for k, v in diffuse.items():
    plt.plot(cloud_cover, v, label=k)
    
plt.legend(); plt.xlabel('cloud cover');
plt.ylabel('I/I0');
```
